```python
def detect_cough(recording, sr):
    """
    Predicts whether a cough is present in the recording using model stored in app.

    Args:
      recording (np.array): user recording.
      sr (int): sample rate
    Returns:
      pred_conf (float): predicted confidence of cough existence.
    """
    features = COUGH_DETECTOR.extract_features(recording, sr)
    # pred_conf = COUGH_DETECTOR.classify_cough(features)[0]
    # logging.info('Cough Detection Prediction:', pred_conf)
    return features


def predict_covid(recording, sr, clinical_feature):
    """Predicts if the cough is healthy, symptomatic (could be any class), or Covid-19.

    Args:
        recording (np.array): cough recording
        sr (int): cough's sample rate
    """
    pred_conf = COVID_CLASSIFIER.classify_cough(recording, sr, clinical_feature)
    # logging.info('Covid Predictions:', pred_conf.tolist())
    return pred_conf
```

```python
import os
from tqdm import tqdm

files_dir = "/home/abrol/ARYAN_MT22019/CoughVid/public_dataset/"
data_vec = []
Sampling_rate_vec = []  # sample rate
features = []
fname = []
status = []
cough_detected = []


for audio_file in tqdm(os.listdir(files_dir)):
    if audio_file.endswith('.webm'):
        audio_contents = audio_file[:-5]
        fname.append(audio_contents)
        if audio_contents in df1['uuid'].values:
            stat_val = df1.loc[df1['uuid'] == audio_contents, 'status']
            status.append(stat_val.iloc[0])
            cough_val = df1.loc[df1['uuid'] == audio_contents, 'cough_detected']
            cough_detected.append(cough_val.iloc[0])


            audio_path = os.path.join(files_dir, audio_file)
            y, sr = librosa.load(audio_path)
            data_vec.append(y)
            Sampling_rate_vec.append(sr)
            clinical_feature = df1[df1['uuid'] == audio_contents]
            f = predict_covid(y, sr, clinical_feature)
            features.append(f)
```
